```python
#Top Spenders
group = choose.groupby("SN") 
purcoun = group["Price"].sum()  
new = purcoun * len(purcoun)/len(purcoun)
array = pd.Series(new).array


df = choose 
mat = df.groupby("SN").sum() 
r = df.groupby("SN").count()
ID = r["Purchase ID"] 
division = array/ID

  
mat['Purchase Count'] = ID 
mat['Average Purchase Price'] = division
mat["Total Purchase Value"] = array
df = mat 

#display/format
matrix = df.nlargest(5,'Total Purchase Value') 
del matrix['Purchase ID'] 
del matrix['Age'] 
del matrix['Item ID'] 
del matrix['Price'] 
matrix['Average Purchase Price'] = matrix['Average Purchase Price'].map("${:,.3}".format) 
matrix['Total Purchase Value'] = matrix['Total Purchase Value'].map("${:,.6}".format)

matrix


```

```python
#Most Popular Items 
group = choose.groupby(["Item ID", "Item Name"]) 
purcoun = group["Price"].sum()  
new = purcoun * len(purcoun)/len(purcoun)
array = pd.Series(new).array

df = choose
v = choice.groupby(["SN"]).count()
mat = df.groupby([ "Item ID","Item Name"]).sum()
r = df.groupby(["Item ID", "Item Name"]).count() 

ID = r["Purchase ID"] 
division = array/ID

mat['Purchase Count'] = ID 
mat['Item Price'] = division
mat["Total Purchase Value"] = array
df = mat 
#format/display 
matrix = df.sort_values(by='Purchase Count', ascending=False)
del matrix['Age']  
del matrix['Price'] 
del matrix['Purchase ID']  
matrix['Item Price'] = matrix['Item Price'].map("${:,.3}".format) 
matrix['Total Purchase Value'] = matrix['Total Purchase Value'].map("${:,.6}".format)
matrix.head()

```

```python
df = matrix
df = df.sort_values('Total Purchase Value')  
df['Total Purchase Value'] = df['Total Purchase Value'].str.replace('$', '')
df['Total Purchase Value'] = df['Total Purchase Value'].astype('float')
df = df.sort_values(by=['Total Purchase Value'], ascending=(False)) 
df['Total Purchase Value'] = df['Total Purchase Value'].map("${:,.7}".format)
df.head()
```
